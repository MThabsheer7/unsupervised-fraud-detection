```python
# Analyzing the V-features which are PCs.
# PCA is a dimensionality reduction technique, so original features are used to create a PC.
# Example: V1 might be 0.3 * original_feature_A + 0.5 * original_feature_B - 0.1 * original_feature_C, and so on.

# Decorrelation: The V-features (V1-V28) are orthogonal (uncorrelated) to each other.

# Lets understand their statistical properties and how they relate to the target variable (Class) and to each other.
```

```python
# STatistical summary
df[['V' + str(i) for i in range(1, 29)]].describe()

# means are very close to zero and standard deviations are around 1, indicating they've been scaled (often implicitly during PCA itself, or explicitly beforehand).
```

```python
skewness = df[['V' + str(i) for i in range(1, 29)]].skew()
sns.pointplot(skewness)
```

```python
kurtosis = df[['V' + str(i) for i in range(1, 29)]].kurt()
sns.pointplot(kurtosis)
```

```python
df[['V' + str(i) for i in range(1, 29)]].corr()
# confirmed correlation values are very close to zero(orthogonal)
```

```python
# Even though you can't interpret them, how the distribution of a V-feature differs between fraudulent and legitimate transactions can be highly insightful for anomaly detection. Fraudulent transactions are "anomalies," and they should look different in some feature dimensions.

# Relationship with Class feature

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Class'] == 0]['V14'], label='Class 0 (Legitimate)', fill=True, alpha=0.5)
sns.kdeplot(df[df['Class'] == 1]['V14'], label='Class 1 (Fraud)', fill=True, alpha=0.5)
plt.title('Distribution of V14 by Class', fontsize=16)
plt.xlabel('V14', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
# plt.savefig('v4_class_distribution.png')
```

```python
# Correlation with Class feature
df.corr()['Class'].sort_values(ascending=False)
```

### Findings
- They are PCA-transformed features for confidentiality and dimensionality reduction.
- Their original meaning is unknown.
- Despite being uninterpretable in original terms, their distributions and correlations with the fraud class clearly differentiate fraudulent from legitimate transactions, making them valuable for anomaly detection.
